```python
import pandas as pd
import numpy as np
import lda
import keras
import tensorflow as tf
import gensim
from gensim import corpora
from gensim.corpora.dictionary import Dictionary
from konlpy.tag import Okt
t = Okt()
```

```python
data = pd.read_csv('/home/ekim_reverse/PyCode/study/Python.study/toy_project/online_news_test.csv')
vocab = pd.read_csv('/home/ekim_reverse/PyCode/study/Python.study/toy_project/stop_words.csv', encoding = 'cp949')
```

```python
for line in data['content']:        #형태소분석기
    news_word = t.pos(line)
```

```python
news_word[0]
```

```python
result_word = []              # 명사, 동사, 형용사만 추출
target = ["Noun","Verb","Adjective"]

for word in news_word:
    if word[1] in target:
        result_word.append(word)
```

```python
result_word
```

```python
result = []        #불용어 제거
for w in result_word:
    if w[0] not in vocab:
        result.append(w)

print(result)
```

```python
result_pick = [i[0] for i in result]     #사용할 단어들만 추출
```

```python
result_pick
```

```python
dictionary = corpora.Dictionary(result)
```

```python
print(dictionary.token2id)           #단어들에 id 부여
```

```python
corpus = [dictionary.doc2bow(result_pick) for result_pick in result]
```

```python
print(dictionary)
```

```python
dictionary[0] 
```

```python
len(dictionary)   # 출력된 수의 단어가 학습
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 10, id2word=dictionary, passes=15)
```

```python
topics = ldamodel.print_topics(num_words=10)
```

```python
for topic in topics:
    print(topic)
```
