```python
from tqdm import tqdm

df = pd.read_csv("/home/ekim_reverse/PyCode/study/Python.study/toy_project/erai_01.csv")
#df.head()

dateList = df["New_date"].unique().tolist()

strList = []

for date in tqdm(dateList):
    
    temp_str = ""
    
    for i in tqdm(range(len(df))):
    
        if date == df["New_date"][i]:
            temp_str = temp_str + "," + str(df["0"][i])
            
    strList.append(temp_str)

df2 = pd.DataFrame({"date": dateList, "str": strList})
```
