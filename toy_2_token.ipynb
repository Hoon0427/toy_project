```python
# 2년치 파일 로드

import pandas as pd 
data=pd.read_excel('/home/ekim_reverse/PyCode/study/Python.study/toy_project/online_news_(1).xlsx')
```

```python
# 형태소 분석기 로드

import nltk
from konlpy.tag import Okt; t = Okt()

pos_tagger = Okt()
```

```python
# Adjective만 추출하는 함수

def find_text_n(x):
    temp=t.pos(x)
    
    data=[]
    
    for line in temp:
        if len(line[0])>1: #한글자 단어는 빼주는게 좋다. 
            if line[1] in ['Noun']: # 명사, 동사, 형용사 라면 
                data.append(line[0])
            
            
    return data
```

```python
# content에서의 형태소 분석 -> Noun 추출

data_pos_t_n=[]
for i in range(len(data)):
    tem=find_text_n(data['title'][i])
    for word in tem: data_pos_t_n.append(word)
```
